Identify the correlation between age and insurance premiums.

#### âœ…Output
As age increases, the average premium also increases.


```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read insurance.csv and store it in df

df=pd.read_csv('insurance.csv')
df.head()
```

```python
# Using the pandas corr() function
correlation = df['age'].corr(df['charges'])
print(f"Correlation coefficient between age and charges: {correlation}")
```

```python
plt.figure(figsize=(10, 8))
sns.regplot(x='age', y='charges', data=df)
plt.title('Relation between age and insurance cost')
plt.xlabel('age')
plt.ylabel('insurance cost')
plt.show()
```
