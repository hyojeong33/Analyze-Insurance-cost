```python
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read insurance.csv and store it in df

df=pd.read_csv('insurance.csv')
```

```python
df['region'].unique()
```

```python
southwest_charges = df[df['region'] == 'southwest']['charges']
southeast_charges = df[df['region'] == 'southeast']['charges']
northwest_charges = df[df['region'] == 'northwest']['charges']
northeast_charges = df[df['region'] == 'northeast']['charges']
```

```python
# Perform one-way ANOVA
f_statistic, p_value = stats.f_oneway(
    southwest_charges,
    southeast_charges,
    northwest_charges,
    northeast_charges
)

print(f"F-Statistics: {f_statistic}")
print(f"P-value: {p_value}")
if p_value<0.05:
    print('The average difference in insurance premiums by region as statistically significant.')
```

```python
regional_stats = df.groupby('region')['charges'].agg(['mean', 'std']).reset_index()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Gets the file 'insurance.csv'.
df = pd.read_csv('insurance.csv')

# 2. Draw a box plot showing the distribution of premiums by region.
plt.figure(figsize=(12, 8))
sns.boxplot(x='region', y='charges', data=df)

# 3. Sets the graph title and axis name.
plt.title('Regional health insurance premium distribution and outlier comparison')
plt.xlabel('Region')
plt.ylabel('Insurance cost ($)')

# 4. Display the graph on the screen.
plt.show()
```
